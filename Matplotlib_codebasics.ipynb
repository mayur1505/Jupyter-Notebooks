### Matplotlib Tutorial 3 - Axes labels, Legend, Grid

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t)
```

```python
#we can all 3 in one chart
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

### Axes labels and chart title

```python
# set axes labels and chart title
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
```

### Legend

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend()
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend(loc='best')#by defalut it is set to best
```

### Matplotlib Tutorial 4 - Bar Chart

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
import numpy as np
```

```python
plt.bar(company,revenue)
```

```python
xpos=np.arange(len(company)) #ypos is yposition
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)#ypos values will be replaced by company names
plt.ylabel('Revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(xpos,revenue,label='Revenue')
plt.legend()
```

#### Multiple Bars showing revenue and profit of major US tech companies

```python
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue)
plt.bar(company,profit)
```

```python
plt.xticks(xpos,company)#xpos values will be replaced by company names
plt.ylabel('Revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(xpos,revenue,label='Revenue')
plt.bar(xpos,profit,label='Profit')
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.ylabel('Revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2,revenue,label='Revenue')
plt.bar(xpos+0.2,profit,label='Profit')
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.ylabel('Revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')
plt.legend()
```

### Horizontal bar chart using barh function

```python
plt.xticks(xpos,company)
plt.ylabel('Revenue(bln)')
plt.title('US Tech Stocks')
plt.barh(xpos-0.2,revenue,label='Revenue')
plt.barh(xpos+0.2,profit,label='Profit')
plt.legend()
```

```python
plt.yticks(xpos,company)
plt.ylabel('Revenue(bln)')
plt.title('US Tech Stocks')
plt.barh(xpos-0.2,revenue,label='Revenue')
plt.barh(xpos+0.2,profit,label='Profit')
plt.legend()
```

```python
plt.yticks(xpos,company)
plt.ylabel('Revenue(bln)')
plt.title('US Tech Stocks')
plt.barh(xpos-0.2,revenue,height=0.4,label='Revenue')
plt.barh(xpos+0.2,profit,height=0.4,label='Profit')
plt.legend()
```

### Matplotlib Tutorial 5 - Histograms

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)
```

These are the y axis numbers ->array([3., 3., 1., 0., 1., 1., 0., 2., 0., 2.])
These are the bins or bucket(10 ranges)->  array([ 77. ,  84.3,  91.6,  98.9, 106.2, 113.5, 120.8, 128.1, 135.4,
        142.7, 150. ]),

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,rwidth=0.5)#rwidth is a relative width of a bar compared to bin size
```

### bins parameter

```python
plt.hist(blood_sugar,rwidth=0.5,bins=4)
```

```python
plt.hist(blood_sugar,bins=3,rwidth=0.95)#rwidth=0.95 means it is like 0.95 of entire range
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g')
```

More info about hist function
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g',histtype='step')
```

### Mutiple data samples in a histogram

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

### horizontal orientation

```python
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, orientation='horizontal')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women']\
        ,orientation='horizontal')
plt.legend()
```

### Matplotlib Tutorial 6 - Pie Chart

#### Draw pie chart to track down home expenses

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
#Tutor got above circle as ellipse so he used plt.axis('equal') to get
#pie chart as perfect circle
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels)
```

```python
#to get rid of all unwanted text..
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

More about pie chart
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

```python
#increasing radius
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=2)
plt.show()
```

```python
#showing percentages
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct='%0.0f%%')
#  %f%% is a format string and 0.0 shows 0 numbers after decimal
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct='%0.2f%%')
#  %f%% is a format string and 0.2 shows 2 numbers after decimal
plt.show()
```

```python
#shadow argument
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct='%0.1f%%',shadow=True)
#  %f%% is a format string and 0.0 shows numbers after decimal
plt.show()
```

### Explode Argument

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,explode=[0,0,0,0,0],radius=1.5,autopct='%0.0f%%')
#  %f%% is a format string and 0.0 shows numbers after decimal
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,explode=[0,0.2,0,0,0],radius=1.5,autopct='%0.2f%%')

plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,explode=[0,0,0,0.2,0.4],radius=1.5,autopct='%0.2f%%')
#0.2 and 0.4 is to instruct that by how much amount pie will come out
#if it is 0 then it will not come out
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,startangle=45,explode=[0,0,0,0.2,0.4],radius=1.5,autopct='%0.2f%%')
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,startangle=180,explode=[0,0,0,0.2,0.4],radius=1.5,autopct='%0.2f%%')
plt.show()
```

```python
#conterclock argument
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,counterclock=False,startangle=180,explode=[0,0,0,0.2,0.4],radius=1.5,autopct='%0.2f%%')
plt.show()
```
